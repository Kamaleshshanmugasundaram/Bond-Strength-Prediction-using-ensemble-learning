```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['(mm)', '(Mpa)', '(Gpa)', '(mm).1', '(mm).2', '(mm).3']]
```

```python
Y=df[['(kN)']]
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X,Y, test_size=0.25)
```

```python
from xgboost import XGBRegressor

xgbr = XGBRegressor()
# Add silent=True to avoid printing out updates with each cycle
xgbr.fit(train_X, train_Y, verbose=False)
```

```python
pred1 = xgbr.predict(test_X)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.inspection import permutation_importance
rmse1 = (np.sqrt(mean_squared_error(test_Y, pred1)))
r21 = r2_score(test_Y, pred1)
print('Testing performance of xgboost regressor')
print('RMSE: {:.2f}'.format(rmse1))
print('R2: {:.2f}'.format(r21))
```

```python
import catboost as cb
```

```python
train_dataset = cb.Pool(train_X, train_Y) 
test_dataset = cb.Pool(test_X, test_Y)
```

```python
cbr = cb.CatBoostRegressor(loss_function='RMSE')
```

```python
grid = {'iterations': [100, 150, 200],
        'learning_rate': [0.03, 0.1],
        'depth': [2, 4, 6, 8],
        'l2_leaf_reg': [0.2, 0.5, 1, 3]}
cbr.grid_search(grid, train_dataset)
```

```python
pred2 = cbr.predict(test_X)
rmse2 = (np.sqrt(mean_squared_error(test_Y, pred2)))
r22 = r2_score(test_Y, pred2)
print('Testing performance of Catboost Regressor')
print('RMSE: {:.2f}'.format(rmse2))
print('R2: {:.2f}'.format(r22))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators = 100, random_state = 0)
```

```python
rfr.fit(train_X,train_Y) 
```

```python
pred3=rfr.predict(test_X)
rmse3 = (np.sqrt(mean_squared_error(test_Y, pred3)))
r23 = r2_score(test_Y, pred3)
print('Testing performance of Random Forest Regressor')
print('RMSE: {:.2f}'.format(rmse3))
print('R2: {:.2f}'.format(r23))
```

```python
pred1 = xgbr.predict(test_X)

from sklearn.metrics import mean_absolute_error
rmse1 = (np.sqrt(mean_squared_error(test_Y, pred1)))
r21 = r2_score(test_Y, pred1)
print("-----------------------------------------------------------------------------------")
print('Testing performance of xgboost regressor')
print('RMSE: {:.2f}'.format(rmse1))
print('R2: {:.2f}'.format(r21))
print("-----------------------------------------------------------------------------------")

pred2 = cbr.predict(test_X)
rmse2 = (np.sqrt(mean_squared_error(test_Y, pred2)))
r22 = r2_score(test_Y, pred2)
print('Testing performance of Catboost Regressor')
print('RMSE: {:.2f}'.format(rmse2))
print('R2: {:.2f}'.format(r22))
print("-----------------------------------------------------------------------------------")

pred3=rfr.predict(test_X)
rmse3 = (np.sqrt(mean_squared_error(test_Y, pred3)))
r23 = r2_score(test_Y, pred3)
print('Testing performance of Random Forest Regressor')
print('RMSE: {:.2f}'.format(rmse3))
print('R2: {:.2f}'.format(r23))
print("-----------------------------------------------------------------------------------")
```
